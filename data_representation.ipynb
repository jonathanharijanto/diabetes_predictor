```python
import pandas as pd

# df == dataframe 
df = pd.read_csv("dataset_diabetes/diabetic_data.csv")
print(df.shape)
```

```python
age_dict = {}
for i in df['age']:
    if i in age_dict:
        age_dict[i] += 1
    else:
        age_dict[i] = 1

print('age\n', age_dict)
```

```python
race_dict = {}
for i in df['race']:
    if i in race_dict:
        race_dict[i] += 1
    else:
        race_dict[i] = 1
print('race\n', race_dict)
```

```python
gender_dict = {}
for i in df['gender']:
    if i in gender_dict:
        gender_dict[i] += 1
    else:
        gender_dict[i] = 1
print('gender\n', gender_dict)
```

```python
print("List of Features that has '?' + the amount of '?':\n")

for col in df.columns:
    if df[col].dtype == object:
        missing_val = df[col][df[col] == '?'].count()
        if missing_val != 0:
            print(col,missing_val)

print('gender', df['gender'][df['gender'] == 'Unknown/Invalid'].count())
```

```python
drugs = ['metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide', 'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone', 'tolazamide', 'examide', 'citoglipton', 'insulin', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone', 'metformin-rosiglitazone', 'metformin-pioglitazone']

for i in drugs:
    a_dict = {}
    for j in df[i]:
        if j in a_dict:
            a_dict[j] += 1
        else:
            a_dict[j] = 1
    print(i, a_dict)
```
